#### 제2중부고속도로 ~ 서울문산고속도로, 
#### 검색일 2020-07-13 ~ 07-19

```python
# 브라우저 생성 및 열기
driver = webdriver.Chrome('C:/chromedriver.exe')

date = [13,14,15,16,17,18,19]
linename = [1010000451,1010000401,1010001001,1010004001,1010004002,1010000501,1010000521,1010001201,1010001101,1010001102,
           1010001103,1010001301,1010000601,1010000172,1010002901] # 수도권제2순환*2, 서울문산 자료 없음
for i in date:
    for j in linename:
        req = requests.get('https://gits.gg.go.kr/web/trafficInfo/webSectionStatistics.do?linename=101&linename2={}'.format(j)+'&lineway=1&linedate=2020-07-{}'.format(i)+'&speedGraph=07')

        soup = BeautifulSoup(req.text,'html.parser')
        data = soup.find("table",{'class': 'type2'})
        table = parser_functions.make2d(data)

        df = pd.DataFrame(data=table[0:],columns=table[0])
        df = df.drop(index=0,axis=0)
        df = df.reset_index(drop=True)
        
        df = df.fillna(0) # 결측값 '일단' 0으로 채움
        
        table_df = df

        print(df)

        table_df.to_excel('webcrawler/linename{}'.format(j) + 'table_df{}'.format(i) +'.xlsx',index=False,encoding='CP949')
        # csv : 1kb, xlsx : 7kb 

        #tables = soup.select('table')
        #table = tables[0]
        #table_html = str(table)
        #globals()["table_df_list{}".format(i)] = pd.read_html(table_html)

        #print(i, globals()["table_df_list{}".format(i)])
```
